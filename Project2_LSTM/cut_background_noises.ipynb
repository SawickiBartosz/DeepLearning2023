# Loading packages

```python
import os
import logging
import shutil

from pydub import AudioSegment
from pydub.utils import make_chunks
```

# Managing files

```python
TRAIN_DATASET_PATH = "tensorflow-speech-recognition-challenge/train/audio"
BACKGROUND_NOISES_PATH = os.path.join(TRAIN_DATASET_PATH, "_background_noise_")
if os.path.exists(os.path.join(TRAIN_DATASET_PATH, "silence")):
    shutil.rmtree(os.path.join(TRAIN_DATASET_PATH, "silence"))
os.mkdir(os.path.join(TRAIN_DATASET_PATH, "silence"))
```

```python
def export_cutted_files(file, output_dir):
    input_file = os.path.join(BACKGROUND_NOISES_PATH, file)
    audio = AudioSegment.from_file(input_file)
    chunks = make_chunks(audio, 1000)
    for i, chunk in enumerate(chunks):
        chunk_name = f"{file[:-4]}_{i}.wav"
        print(f"Exporting {chunk_name}")
        output_file = os.path.join(output_dir, chunk_name)
        chunk.export(output_file, format="wav")
```

```python
for file in os.listdir(BACKGROUND_NOISES_PATH):
    if file.endswith("wav"):
        export_cutted_files(file, os.path.join(TRAIN_DATASET_PATH, "silence"))
```
