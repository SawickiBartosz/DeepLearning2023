```python
import numpy as np
import pandas as pd
import os
```

```python
results = np.load("results/final_result.npy") 
results.shape
```

```python
results[0:5]
```

```python
filenames = list(os.walk(os.path.join(os.getcwd(), '..', 'cifar-10', 'test'), topdown=False))[0][2]
filenames = pd.DataFrame(filenames, columns=["name"])
filenames["id"] = filenames.name.str[:-4].astype(int)
filenames["classes"] = results
filenames
```

```python
map_class = {
    0: "airplane",
    1: "automobile",
    2: "bird",
    3: "cat",
    4: "deer",
    5: "dog",
    6: "frog",
    7: "horse",
    8: "ship",
    9: "truck"
}
filenames["label"] = filenames.classes.map(map_class)
filenames
```

```python
filenames[["id", "label"]].to_csv("submission.csv", index=False)
```
