```python
import pickle
import torch
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

# WGAN

```python
losses = torch.load("GANs/WGAN-GP-v2_512/losses.pt")
```

```python
losses.keys()
```

```python
g = losses["G"]
d = losses["D"]
gp = losses["GP"]
g_norm = losses["gradient_norm"]
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 8), sharex=True)
ax[0][0].plot(d[::5])
ax[0][0].set_title('Critic loss')

ax[0][1].plot(gp[::5])
ax[0][1].set_title('Gradient penalty')

ax[1][0].plot(g)
ax[1][0].set_title('Generator loss')
ax[1][0].set_xlabel('Training iteration')

ax[1][1].plot(g_norm[::5])
ax[1][1].set_title('Gradient norm')
ax[1][1].set_xlabel('Training iteration')
```

# GANs

```python
with open(os.path.join("GANs", "DC-GAN_32", "loss_D.pkl"), 'rb') as f:
    loss_D = pickle.load(f)
with open(os.path.join("GANs", "DC-GAN_32", "loss_G.pkl"), 'rb') as f:
    loss_G = pickle.load(f)

plt.figure(figsize=(10,5))
plt.title("Generator and Discriminator Loss During Training - 32 latent space dimensionality")
plt.plot(loss_G, label="G")
plt.plot(loss_D, label="D")
plt.xlabel("iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
