## Add stylegan3 repository to python path

```python
import os
os.environ['CUBLAS_WORKSPACE_CONFIG'] = ':4096:8'
import torch
import random
from torch.utils.data import DataLoader
from torchvision.transforms import transforms
from torchvision import datasets
from torch import optim
from wgan_gp.training import Trainer
from wgan_gp.models import Generator, Discriminator
```

```python
BATCH_SIZE = 256
IMG_SIZE = (64, 64)
IMG_CHANNELS = 3
LEARNING_RATE = 1e-4
NUM_EPOCHS = 100
manualSeed = 999
print("Random Seed: ", manualSeed)
random.seed(manualSeed)
torch.manual_seed(manualSeed)
torch.cuda.manual_seed(manualSeed)
torch.use_deterministic_algorithms(True)
```

## Train on lsun bedrooms

```python
dataset = datasets.ImageFolder("lsun_bedroom", transform=transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
]))
loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=6)
```

```python
latent_dims = [512]
device = torch.device("cuda:0")

for latent_dim in latent_dims:
    generator = Generator(img_size=(*IMG_SIZE, IMG_CHANNELS), latent_dim=latent_dim, dim=64).to(device)
    discriminator = Discriminator(img_size=(*IMG_SIZE, IMG_CHANNELS), dim=64).to(device)
    # Setup Adam optimizers for both G and D
    optimizerD = optim.Adam(discriminator.parameters(), lr=LEARNING_RATE, betas=(0.9, 0.99))
    optimizerG = optim.Adam(generator.parameters(), lr=LEARNING_RATE, betas=(0.9, 0.999))
    # Create batch of latent vectors that we will use to visualize
    #  the progression of the generator
    trainer = Trainer(generator, discriminator, optimizerG, optimizerD,
                      use_cuda=torch.cuda.is_available(), print_every=100)
    trainer.train(loader, NUM_EPOCHS, save_training_gif=False)
    path = os.path.join("GANs", f"WGAN-GP-v2_{latent_dim}")
    os.makedirs(path, exist_ok=True)
    torch.save(trainer.G.state_dict(), os.path.join(path, 'generator.pt'))
    torch.save(trainer.D.state_dict(), os.path.join(path, 'discriminator.pt'))
    torch.save(trainer.losses, os.path.join(path, 'losses.pt'))
```
